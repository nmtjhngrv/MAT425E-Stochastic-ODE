```python
import nbconvert
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
def g(x):
    a,b,c=7**5,0,2**31-1
    return (a*x+b)%c
a_list=[]
b_list=[]
x=1
for i in range (50):
    a_list=np.append(a_list,x)
    x=g(x)
x=7
for i in range (50):
    b_list=np.append(b_list,x)      
    x=g(x)
```

```python
print(a_list)
plt.title('Pseudo -Random Generators')
plt.plot(a_list)
plt.xlabel('X')
plt.ylabel('y')
plt.figure(figsize=(20,20))
plt.show()
```

```python
print(b_list)
plt.title('Pseudo -Random Generators')
plt.plot(b_list)
plt.xlabel('X')
plt.ylabel('y')
plt.figure(figsize=(20,20))
plt.show()
```
